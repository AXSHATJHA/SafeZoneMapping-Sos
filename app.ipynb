```python
df_proj = df[["State/ UT", "District/ Area", "Total Crimes against Women"]].copy()
```

```python
df_proj
```

```python
state_totals = df_proj.groupby("State/ UT")["Total Crimes against Women"].transform("sum")
```

```python
df_proj["Crime_Probability"] = df_proj["Total Crimes against Women"] / state_totals
```

```python
#Crime Weights for each crime

crime_weights = {
    "Rape": 20,
    "Attempt to commit Rape": 15,
    "Dowry Deaths": 0,
    "Abetment of Suicides of Women": 16,
    "Kidnapping & Abduction_Total": 18,
    "Cruelty by Husband or his Relatives": 10,
    "Assault on Women with intent to outrage her Modesty_Total": 14,
    "Insult to the Modesty of Women_Total": 10,
    "Protection of Children from Sexual Offences Act": 20,
    "Importation of Girls from Foreign Country": 17,
    "Immoral Traffic Prevention Act": 0,
    "Dowry Prohibition Act, 1961": 0,
    "Protection of Women from Domestic Violence Act, 2005": 10,
    "Indecent Representation of Women (P) Act, 1986": 0
}
```

```python
df_proj["District_Score"] = sum(df[col] * weight for col, weight in crime_weights.items())
```

```python
df_proj
```

```python
state_totals = df_proj.groupby("State/ UT")["District_Score"].transform("sum")
```

```python
df_proj["Normalized_Score"] = df_proj["District_Score"] / state_totals
```
